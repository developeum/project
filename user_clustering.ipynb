```python
import pandas as pd
from sklearn.cluster import AffinityPropagation
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
user_data = pd.read_csv('data/user_data.csv')
```

```python
user_data = user_data.to_numpy()
```

```python
af = AffinityPropagation(preference=-15, max_iter=1000).fit(user_data)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
```

```python
len(np.unique(labels))
labels
```

```python
pca = PCA(n_components=2).fit(user_data)

X_pca = pca.transform(user_data)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap=plt.cm.tab10)
```

```python
cluster1 = []
for x in labels:
    if labels[x]==0:
        cluster1.append(user_data[x])
```

```python
it_df = pd.read_csv('data/IT-мероприятия_new.csv')
events = []
for x in it_df['Какие из перечисленных мероприятий интересны Вам?'].to_numpy():
    for i in x.split(';'):
        events.append(i)
events = np.unique(events)

cl1 = pd.DataFrame(cluster1,columns=events)
```
